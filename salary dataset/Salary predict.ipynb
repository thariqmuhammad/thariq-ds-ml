```python
import pandas as pd

df = pd.read_csv('train.csv')
df.tail()
```

```python
# Replace gaji
gaji_dict = {
    '<=7jt' : 0,
    '>7jt' : 1
}
df['Gaji'] = df['Gaji'].replace(gaji_dict)
```

```python
# Berat Akhir
df['Berat'] = pd.cut(df['Berat Akhir'].astype(int), 5)
df[['Berat','Gaji']].groupby(['Berat'],as_index=False).mean()
```

```python
# Jam
df['Waktu Kerja'] = pd.cut(df['Jam per Minggu'].astype(int), 4)
df[['Waktu Kerja','Gaji']].groupby(['Waktu Kerja'],as_index=False).mean()
```

```python
# Keuntungan kapital

for row in df:
    df['Untung'] = 0
    df.loc[df['Keuntungan Kapital'] > 1, 'Untung']= 1

df[['Untung','Gaji']].groupby(['Untung'],as_index=False).mean()
```

```python
# Kerugian kapital

for row in df:
    df['Rugi'] = 0
    df.loc[df['Kerugian Capital'] > 1, 'Rugi']= 1

df[['Rugi','Gaji']].groupby(['Rugi'],as_index=False).mean()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

df['Status'] = label.fit_transform(df['Status Perkawinan']) 
df['Pendidikan_Code'] = label.fit_transform(df['Pendidikan']) 
df['Pekerjaan_Code'] = label.fit_transform(df['Pekerjaan']) 
df['Gender'] = label.fit_transform(df['Jenis Kelamin']) 
#df['Umur_Code'] = label.fit_transform(df['Kategori Umur']) 
df['Berat_Code'] = label.fit_transform(df['Berat']) 
df['Jam Kerja_Code'] = label.fit_transform(df['Waktu Kerja']) 
```

```python
drop_dict = ['id', 'Kelas Pekerja', 'Berat Akhir','Keuntungan Kapital', 'Kerugian Capital','Jam per Minggu',
             'Pendidikan','Pekerjaan','Jenis Kelamin','Berat','Waktu Kerja','Status Perkawinan']
df_result_train = df.drop(drop_dict, axis=1)
df_result_train.info()
```

# Test

```python
import pandas as pd

df_test = pd.read_csv('test.csv')
df_test
```

```python
# Berat Akhir
df_test['Berat'] = pd.cut(df_test['Berat Akhir'].astype(int), 5)
```

```python
# Jam
df_test['Waktu Kerja'] = pd.cut(df_test['Jam per Minggu'].astype(int), 4)
```

```python
# Keuntungan kapital

for row in df_test:
    df_test['Untung'] = 0
    df_test.loc[df_test['Keuntungan Kapital'] > 1, 'Untung']= 1
```

```python
# Kerugian kapital

for row in df_test:
    df_test['Rugi'] = 0
    df_test.loc[df_test['Kerugian Capital'] > 1, 'Rugi']= 1
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

df_test['Status'] = label.fit_transform(df_test['Status Perkawinan']) 
df_test['Pendidikan_Code'] = label.fit_transform(df_test['Pendidikan']) 
df_test['Pekerjaan_Code'] = label.fit_transform(df_test['Pekerjaan']) 
df_test['Gender'] = label.fit_transform(df_test['Jenis Kelamin']) 
#df_test['Umur_Code'] = label.fit_transform(df_test['Kategori Umur']) 
df_test['Berat_Code'] = label.fit_transform(df_test['Berat']) 
df_test['Jam Kerja_Code'] = label.fit_transform(df_test['Waktu Kerja']) 
```

```python
drop_dict = ['id', 'Kelas Pekerja', 'Berat Akhir','Keuntungan Kapital', 'Kerugian Capital','Jam per Minggu',
             'Pendidikan','Pekerjaan','Jenis Kelamin','Berat','Waktu Kerja','Status Perkawinan']
df_result_test = df_test.drop(drop_dict, axis=1)
df_result_test
```

# Modelling

```python
# Modelling
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score, cross_validate
import pandas as pd
import numpy as np

# Processing data
x_train = df_result_train.drop('Gaji', axis=1)
y_train = df_result_train['Gaji']

x_test = df_result_test

```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# membuat model
knn = KNeighborsClassifier()
param_grid = {'n_neighbors' : np.arange(1, 50), 'weights' : ['uniform', 'distance']}
gscv = GridSearchCV(knn, param_grid=param_grid, scoring = 'roc_auc', cv=10)

# training data
gscv.fit(x_train,y_train)
```

```python
# hyperparameter terbaik untuk kombinasi
gscv.best_params_
```

```python
# score validasi terbaik
gscv.best_score_
```

```python
# Prediksi probabilitas
probabilitas = gscv.predict_proba(x_test)
probabilitas
```

```python
# Prediksi data test
y_pred = gscv.predict(x_test)
y_pred
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

model = DecisionTreeClassifier()
param_dist = {"max_depth": [50, None],
              "min_samples_leaf": randint(1, 51),
              "criterion": ["gini", "entropy"]}

rscv = RandomizedSearchCV(model, param_dist, scoring = 'roc_auc', cv=10)

# Latih model terhadap data training
rscv.fit(x_train,y_train)
```

```python
# Prediksi data test
y_pred = rscv.predict(x_test)
y_pred
```
